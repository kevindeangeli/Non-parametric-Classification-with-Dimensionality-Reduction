```python


def euclidian_dsitance(x,X):
    #x is the test point, X is the dataset
    #Using Euclidian Distance
    distancesArr = []
    #For each row in the data set:
    dist=-1
    for index,row in X.iterrows(): 
        testPoint= row.loc[0:X.shape[1]-2]
        dist = np.linalg.norm(testPoint-x)
        label= X.loc[index][X.shape[1]-1]
        distanceAndLabel = (dist**(1/2),label)
        distancesArr.append(distanceAndLabel)
    return distancesArr

    
    
def guessLabel(x,X,k):
    label0 =0
    guessClass = 9 
    label1 =0
    ks = []
    distancesArr=euclidian_dsitance(x,X)
    distancesArr.sort()
    for p in range(int(k)):
        minimum= min(distancesArr)
        ks.append((minimum[0],minimum[1]))
        distancesArr.remove(minimum)
    for q in range(len(ks)):
        if ks[q][1]==0:
            label0 += 1
        else:
            label1 += 1
    if label0>= label1:
        guessClass = 0
    else:
        guessClass = 1
    return guessClass


    
def knn(testX,X,k,displayStats = True, priors=False,alpha=-1):
#Here testX is the test data set
#X is the trainning data set
#K is the number of neighbors 
#let label 1 represent True
#and Label 0 represent False
    start_time = time.time()
    TP,TN,FN,FP =0,0,0,0
    for index,row in testX.iterrows(): 
        trueClass = testX.loc[index,testX.shape[1]-1]
        testPoint = testX.loc[index,0:X.shape[1]-2]
        if priors == True:
            predictedClass=guessLabelWithPriors(testPoint,X,k,alpha)
        else:
            predictedClass = guessLabel(testPoint,X,k)
        if trueClass == predictedClass:
            if trueClass == 1:
                TP +=1
            else:
                TN +=1
        else:
            if trueClass == 0:
                FP +=1 #Since it was guess as 1.
            else:
                FN +=1
    accuracy = (TP+TN)/ testX.shape[0]
    if displayStats == True:
        print("K = ", k)
        print("TP: ", TP, "TN: ", TN, "FP: ", FP, "FN: ", FN )
        print("accuracy: ", accuracy*100)
        print("------------------------")
        print("--- %s seconds ---" % (time.time() - start_time))
    else:
        accuracyList = [accuracy, TP,TN,FP,FN]
        return accuracyList


    
```

```python
def ConfusionMatrixWithPriors(testData,case):
    #g0,g1 = case1()
    #Let class 0 = N; Class 1 = P;
    TP,TN,FP,FN = 0,0,0,0
    for index,row in testData.iterrows():
        x1= row[0:testData.shape[1]-1]
        
        g0_out = case.g0(x1,case.data.w0)
        g1_out = case.g1(x1,case.data.w1)
        if g0_out>=g1_out:
            guessLabel=0
        else:
            guessLabel=1
        rightLabel=row[testData.shape[1]-1]
        if guessLabel==rightLabel:
            if rightLabel == 1:
                TP+=1
            else:
                TN+=1
        else:
            if rightLabel==1:
                FN+=1
            else:
                FP+=1
    #print(TP)
    totalRowsInData = nX_test.shape[0]
    confusion_matrix = [['TP', TP], ["TN", TN], ['FP', FP], ['FN', FN],['Accuracy', (TN+TP)/totalRowsInData]]
    confusionarr = [TP, TN, FP, FN]
    return confusion_matrix, confusionarr
```

```python
#given a dataset, this class stores all the relevant variables
class GaussianStatistics():
    def __init__(self, X):
        self.y1Values =  X[X[X.shape[1]-1]==1]
        self.y0Values =  X[X[X.shape[1]-1]==0]

        self.y0Values=self.y0Values.drop(self.y0Values.columns[self.y0Values.shape[1]-1], axis=1)
        self.y1Values=self.y1Values.drop(self.y1Values.columns[self.y1Values.shape[1]-1], axis=1)

        self.y0meanArr = np.mean(self.y0Values, axis=0)
        self.y1meanArr = np.mean(self.y1Values, axis=0)

        self.y0Cov = self.y0Values.cov()
        self.y1Cov= self.y1Values.cov()

        self.y0Var = self.y0Values.var()
        self.y1Var = self.y1Values.var()

        self.y0Std = self.y0Values.std()
        self.y1Std = self.y1Values.std()

        self.w0 = (self.y0Values.shape[0])/X.shape[0]
        self.w1 = 1-self.w0
```

```python
#*******************************************************************************************
#Case I
#Testing the performance of the Case I Classifier:
#*******************************************************************************************

class case1:
    def __init__(self, X):
        self.data = GaussianStatistics(X)

        self.simgaY0 = (np.average(self.data.y0Std)+np.average(self.data.y1Std))/2

        self.mewY0 = self.data.y0meanArr
        self.mewY1 = self.data.y1meanArr

    def g0(self,x, w):
        V= x
        return (((-1/(2*self.simgaY0)) * (np.dot(V,np.transpose(V)) \
        -2*np.dot(self.mewY0,np.transpose(V))+ np.dot(self.mewY0,np.transpose(self.mewY0))))+math.log(w)).item()

    def g1(self,x, w):
        V= x
        return (((-1/(2*self.simgaY0)) * (np.dot(V,np.transpose(V)) \
        -2*np.dot(self.mewY1,np.transpose(V))+ np.dot(self.mewY1,np.transpose(self.mewY1))))+math.log(1-w)).item()


```

```python
#*******************************************************************************************
#Case II
#Discriminat Function 
#Evaluating the acuracy.
#*******************************************************************************************
class case2:
    def __init__(self, X):
        self.data = GaussianStatistics(X)


        self.Ex = np.zeros((self.data.y1Values.shape[1],self.data.y1Values.shape[1]))

        for i in range(self.data.y1Values.shape[1]):
            self.Ex[i][i] = (self.data.y0Std[i]+self.data.y1Std[i])/2
    

        self.Ex_inv = np.linalg.inv(self.Ex)

        self.u0 = self.data.y0meanArr
        self.u1 = self.data.y1meanArr

    def g0(self,X, w):
        LHS1= np.dot(np.dot(self.Ex_inv,self.u0),X)
        LHS2= -.5 * np.dot(np.dot(self.u0,self.Ex_inv),self.u0)
        LHS3= math.log(w)
        LHS = LHS1 + LHS2 + LHS3
        return LHS

    def g1(self,X, w):
        RHS1= np.dot(np.dot(self.Ex_inv,self.u1),X)
        RHS2= -.5 * np.dot(np.dot(self.u1,self.Ex_inv),self.u1)
        RHS3= math.log(1-w)
        RHS = RHS1 + RHS2 + RHS3
        return RHS


#ConfusionMatrixWithPriors(nX_test,g0,g1,w0,w1)
```

```python
#*******************************************************************************************
#Case III
#Evaluating the classifiers:
#*******************************************************************************************
class case3:
    def __init__(self, X):
        self.data = GaussianStatistics(X)
        self.E0= self.data.y0Cov
        self.E1 = self.data.y1Cov
        self.E0_inv = np.linalg.inv(self.E0)
        self.E1_inv = np.linalg.inv(self.E1)
        self.u0 = self.data.y0meanArr
        self.u1 = self.data.y1meanArr

    def g0(self,X, w):
        LHS1A= -.5*self.E0_inv
        LHS1 = np.dot(np.dot(np.transpose(X),LHS1A),X)
        LHS2 = np.dot(np.transpose(np.dot(self.E0_inv,self.u0)),X)
        LHS3A= np.dot(np.dot(np.transpose(self.u0),self.E0_inv),self.u0)
        LHS3B= np.log(np.linalg.det(self.E0))
        LHS3 = -.5*(LHS3A+LHS3B)
        LHS = LHS1 + LHS2 + LHS3 + math.log(w)
        return LHS

    def g1(self,X, w):
        RHS1A= -.5*self.E1_inv
        RHS1 = np.dot(np.dot(np.transpose(X),RHS1A),X)
        RHS2 = np.dot(np.transpose(np.dot(self.E1_inv,self.u1)),X)
        RHS3A= np.dot(np.dot(np.transpose(self.u1),self.E1_inv),self.u1)
        RHS3B= np.log(np.linalg.det(self.E1))
        RHS3 = -.5*(RHS3A+RHS3B)
        RHS = RHS1 + RHS2 + RHS3 + math.log(1-w)
        return RHS

#ConfusionMatrixWithPriors(nX_test,g0,g1,w0,w1)
```

```python
def plotSensitivity_Specificity(case,testData,trainning_data):
    plt.figure(num=None, figsize=(8, 8), dpi=100, facecolor='w', edgecolor='k')
    x = np.linspace(0.01,1,10)
    y1Sens =[]
    y2Spec=[]
    
    for i in x:
        case.data.w0=i
        ignore, results = ConfusionMatrixWithPriors(testData,case)
        sensitivity = results[0] / (results[0] + results[3])
        specificity = results[1] / (results[1] + results[2])
        y1Sens.append(sensitivity)
        y2Spec.append(specificity)
    plt.plot(x,y1Sens,label='Sensitivity')
    plt.plot(x,y2Spec,label="Specificity")
    plt.legend(loc = 'best')
    plt.xlabel('Prior Probability, w0')
    plt.ylabel('TP/(TP+FN) , TN/(TN+FP)')

```
